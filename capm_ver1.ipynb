```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
data1 = pd.read_csv('tickers.csv')
data2 = pd.read_excel('MCAP_TOP100.xlsx', )
data3 = pd.read_csv('MCAP_TOP100BSE.csv')
```

```python
print(data2.columns.ravel())
print(data3.columns.ravel())
top100nse=data2['Symbol'].tolist()
print(top100nse)
top100bse=data3['Symbol'].tolist()
print(top100bse)
```

```python
a=data1['Ticker'].unique()
```

```python
from difflib import get_close_matches
matches=pd.DataFrame()
for i in top100nse: 
    temp=get_close_matches(i, a, 2)
    matches=matches.append(temp)
matches.reset_index(level=0, inplace=True)
matches.columns.values[1] = 'Name'
print(matches)
```

```python
bse = matches.loc[matches['index'] == 1]
bselist = bse['Name'].tolist() 
print(bselist)
nse = matches.loc[matches['index'] == 0]
nselist = nse['Name'].tolist() 
print(nselist)
```

```python
stock_stat=pd.DataFrame()
names = pd.Series(nselist)
for stock in nselist:
    df1 = yf.download(stock,'2010-08-01','2020-08-01')
    df1['Stock Name']= stock
    stock_stat=stock_stat.append(df1)
stock_price=stock_stat.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1)
stock_price.reset_index(level=0, inplace=True)
#print(stock_price)
#stock_price.to_excel('stock_price.xlsx')
```

```python
market_stat=pd.DataFrame()
for stock in d:
    dfm = yf.download(stock,'2010-08-01','2020-08-01')
    dfm['Stock Name']= stock
    market_stat=market_stat.append(dfm)
market_price=market_stat.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1)
market_price.reset_index(level=0, inplace=True)
print(market_price)
#market_price.to_excel('market_price.xlsx')
```

```python
market_price.info()
```

```python
common = market_price.merge(stock_price,on=['Date'])
common = common.sort_values(by=['Stock Name_y', 'Date'])
print(common)
#common.to_excel('common.xlsx')
```

```python
common['pct_change_stock'] = common.groupby(['Stock Name_y'])['Adj Close_y'].pct_change()
common['pct_change_market']= common.groupby(['Stock Name_y'])['Adj Close_x'].pct_change()
print(common)
#common.to_excel('pct.xlsx')
```

```python
num = common.groupby(['Stock Name_y'])['pct_change_market'].cov(common['pct_change_stock'])
den = common.groupby(['Stock Name_y'])['pct_change_market'].var()
beta = num/den
print(beta)
```

```python
beta_frame=beta.to_frame()
beta_frame.reset_index(level=0, inplace=True)
beta_sorted=beta_frame.sort_values(by=['pct_change_market'])
beta_sorted.rename(columns = {'pct_change_market':'Beta'}, inplace = True) 
print(beta_sorted)
#beta_sorted.to_excel('beta.xlsx')
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20, 7
beta_sorted.plot(x='Stock Name_y',y='Beta',color='red', mfc='black',marker='x', kind='line', fontsize=12)
plt.figure(figsize=(20,10))
plt.show()
```

```python
stock_stat2=pd.DataFrame()
for stock in nselist:
    df2 = yf.download(stock,'2013-08-01','2018-08-01')
    df2['Stock Name']= stock
    stock_stat2=stock_stat2.append(df2)
stock_price2=stock_stat2.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1)
stock_price2.reset_index(level=0, inplace=True)
#print(stock_price2)


common2 = market_price.merge(stock_price2,on=['Date'])
common2 = common2.sort_values(by=['Stock Name_y', 'Date'])
#print(common2)
#common2.to_excel('common2.xlsx')

common2['pct_change_stock'] = common2.groupby(['Stock Name_y'])['Adj Close_y'].pct_change()
common2['pct_change_market']= common2.groupby(['Stock Name_y'])['Adj Close_x'].pct_change()
#print(common2)
#common2.to_excel('pct2.xlsx')

num2 = common2.groupby(['Stock Name_y'])['pct_change_market'].cov(common2['pct_change_stock'])
den2 = common2.groupby(['Stock Name_y'])['pct_change_market'].var()
beta2 = num2/den2
#print(beta2)

beta_frame2=beta2.to_frame()
beta_frame2.reset_index(level=0, inplace=True)
beta_pre2018=beta_frame2.sort_values(by=['pct_change_market'])
beta_pre2018.rename(columns = {'pct_change_market':'Beta'}, inplace = True) 
print(beta_pre2018)
beta_pre2018.to_excel('beta2.xlsx')
```
